# The Monty Hall problem

This is an interactive game and simulator of the Monty Hall problem.

There are 3 doors in front of you, and there is a prize behind one of them.

Once you select a door, I will open one of the two you had not selected which does not have a prize behind it.

You will then have the opportunity to switch from the door you originally selected to an alternate door...should you switch?

```python
import random
A = "A"
B = "B"
C = "C"
```

Monty's Door: Function to choose the door to reveal

```python
def montys_choice(doors, prize, selection):
    if selection == prize:
            open_door = random.choice(list(set(doors) - set(selection)))
    else:
            open_door = random.choice(list(set(doors) - set(selection) - set(prize)))
    return(open_door)
```

Function to determine whether you win

```python
def win_lose(switch_choice, selection):
    if (switch_choice == "N" and selection == prize) or (switch_choice == "Y" and selection != prize):
        outcome = ("Win")
    else:
        outcome = ("Lose")
    return(outcome)
```

## Interactive game starts here:

```python
doors = ["A", "B", "C"]
prize = random.choice(doors)
selection = input("Select door 'A', 'B', or 'C': ")
```

```python
open_door = montys_choice(doors, prize, selection)
print ("The door you chose is: %r" % selection)
print ("The door I will now open is: %r"% open_door)
```

```python
switch_choice = input ("Would you like to change your choice? Type 'Y' or 'N':")
```

Drum roll: Did you win?

```python
outcome = win_lose(switch_choice, selection)    
print ("You %r" %  outcome)
print ("Prize: %r" % prize)
print ("Initial selection: %r " % selection)
print ("Did you switch: %r " % switch_choice)
```

## So should you switch?  Here's a simulation to check your intuition...

Repeating the process 10000 times, with random choices of initial selection, Monty's choice and switching...counting wins when you switch and stick.

```python
switch_choices = ["Y", "N"]
switch_wins = 0
stick_wins = 0
for i in range(10000):
    doors = ["A", "B", "C"]
    prize = random.choice(doors)
    print ("Prize: %r" % prize)
    selection = random.choice(doors)
    print ("Initial selection: %r " % selection)
    open_door = montys_choice(doors, prize, selection)
    print ("The door I will now open is: %r"% open_door)
    switch_choice = random.choice(switch_choices)
    print ("Did you switch: %r " % switch_choice)
    print ("Outcome: %r" % win_lose(switch_choice, selection))
    if win_lose(switch_choice, selection) == "Win":
        if (switch_choice == "Y"):
            switch_wins = switch_wins + 1
        else:
            stick_wins = stick_wins + 1
```

## The big reveal: How often did you win by switching vs sticking?

```python
print ("Number of wins with switch: %r" % switch_wins)
print ("Number of wins witout switch: %r" % stick_wins)
print ("Ratio of switch wins to stick wins: %r" % (switch_wins/stick_wins))
```

Before Monty's choice was revealed, you had a 1/3 chance of winning - on a frequentist basis. 

After Monty's choice you are about twice as likely to win if you SWITCH, winning about 1/3 of the time but only 1/6 of the time when you STICK!  

How can revealing information reduce your chances of winning?
